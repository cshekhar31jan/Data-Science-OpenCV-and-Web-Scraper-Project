# Movie data Visualization :

```python
movie = pd.read_csv("./movie_meta.csv")
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie['movie_title'][2]
```

We can see that, there is some extra characters at the end of each movie_title (not only 1, more than 1 char also), so we first need to remove this extra chars just in case if they may cause some hinderance in some future function :-

lstrip() removes the extra white spaces from the left side of the elements and similiarly rstrip().
strip() removes from bothends.

```python
print(len(movie['movie_title'][2]))
print(movie['movie_title'][2])
```

```python
def preProcess(ele):
    return len(ele.strip())
```

```python
movie['movie_title_len'] = movie['movie_title'].apply(preProcess)
print(len(movie['movie_title'][2]))
print(movie['movie_title'][2])
print(movie['movie_title_len'][2])
```

```python
# It is important to check and run the below code and understand why the histogram is like this.
# plt.hist(len(movie['movie_title']), bins = 50, color= 'blue')
```

```python

plt.hist(movie['movie_title_len'], bins = 50, color= 'blue')
plt.xlabel("movie_title")
plt.xticks(range(0,100,5))
plt.ylabel("correct_length_of_names")
plt.show()
```

### Scatter Way 

```python
movie['movie_title_len']
```

```python
freq_title = {}
for i in movie['movie_title_len']:
    if freq_title.get(i) is None:
        freq_title[i]=1
    else:
        freq_title[i] += 1
```

```python
#This shows the frequency of each movie_len
freq_title
```

```python
freq_title.keys()
```

```python
x = freq_title.keys()
y = freq_title.values()
```

```python
plt.scatter(x, y)
plt.xticks(range(0,100,10))
plt.xlabel("movie title len")
plt.ylabel("Frequency")
plt.show()
```
