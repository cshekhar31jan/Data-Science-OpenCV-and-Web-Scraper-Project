## Now , Lets create the csv file from the collected informations :

So, First we will create a dataframe and from data frame we will create the csv file, so

```python
import pandas as pd
```

```python
# Lets create a dataframe:
df = pd.DataFrame({}, columns= ['Country', 'Capital', 'Language'])
df
```

```python
for one_table in all_tables:
    all_rows =one_table.find_all('tr')[1:]
    
    for tr in all_rows:
        all_tds = tr.find_all('td') 
        #print(all_tds)
        #print(len(all_tds))
        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text
        
        ele= {'Country' : country, 'Capital' : capital, 'Language' : language}
        # Mistake := df.append(ele)
        # correct :-
        df = df.append(ele, ignore_index= True)
    
```

```python
df.shape
```

```python
df.head()
```

You can notice the slash in the language column so,To remove this :

```python
df['Language'][0][-1]
```

```python
# So, collect all the informatin except the last "\n" character :
df['Language'][0][:-1]
```

```python
#Now,Apply For all the values :
df['Language'] = df['Language'].apply(lambda x = x[:-1])
df.head()
```

```python
df.to_csv("countries.csv", index = False)
```

### Now, You will notice that there is an extra space before name of each countries in the csv file, so,Now lets remove that :

Refining The CSV file :

```python
# Notice the extra space:
df['Country'][3]
```

```python
df['Country'].apply(lambda x: x .strip())[3]
```

But still you can see these extra things like [1] at the end of name of American Samoa, So, to remove them , use a library 'regular expressions' ad can be imported as simply 're:

```python
import re
```

RE basically substitutes your expression with the pattern which you will give as arguements :

```python
# For example :
str = "keepcoding[2]"
print(str)
str = re.sub('\[.*\]',"",str)
print(str)
```

```python
df['Country'].apply(lambda x : re.sub('\[.*\]',"",x.strip()))
```
