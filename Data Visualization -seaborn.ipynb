```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Better way of reading data
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#sns.dstplot() Plot the distribution of the data like a histogram
sns.distplot(iris['petal_length'], bins = 20)
# sns.histplot(iris['petal_length'], bins = 20)
plt.show()
```

FutureWarning: `distplot` is a deprecated function and will be removed in a future version. Please adapt your code to use either `displot` (a figure-level function with similar flexibility) or `histplot` (an axes-level function for histograms).

```python
#To get only KDE graph :
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x = iris['petal_length'], y=iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x = 'petal_length', y='petal_width' , data = iris, hue = 'species')
# Same response will be there
# plt.show()
```

```python
sns.jointplot(x = 'sepal_width', y = 'sepal_length', data= iris)
plt.show()
```

```python

sns.jointplot(x = 'sepal_width', y = 'sepal_length', data= iris, kind = 'hex')
plt.show()
```

```python
# To show regression line
sns.jointplot(x = 'sepal_width', y = 'sepal_length', data= iris, kind = 'reg')
plt.show()
```

Pair Plot will create all the different data plot from the dataframe given

-> On diagonal line, you can see (distplot) or (histplot).

```python
sns.pairplot(data  = iris)
```

```python
sns.pairplot(data=iris, hue = 'species')
```

### Seaborn categorical Plot :

```python
sns.countplot(x = 'species', data = iris)
plt.show()
```

```python
sns.boxplot(x = 'species', y = 'petal_width', data = iris)
plt.show()
```

```python
sns.boxplot(x = 'species', y = 'petal_length', data = iris)
plt.show()
```

-> The middle line in each box of above plot shows the mean value of that category

-> box represents the 25-75 % of the data of that category

-> There are some extra point (diamond type shape), called outlier which shows some exp=ceptional behaviors.

```python
sns.violinplot(x = 'species', y='petal_length', data= iris )
plt.show()
```

You should notice the small box plot inside the voilin plot.

```python
sns.violinplot(x = 'species', y='sepal_length', data = iris)
plt.show()
```

# Visualizing Tips Dataset :-
It comes along with seaborn library.

```python
tips = sns.load_dataset('tips')
```

Lets try to get some insgight from data :

```python
sns.countplot(x = tips['sex'])
plt.show()
```

```python
tips['day'].unique()
```

```python
# Out of all the days, which day we get th ehighest amount
sns.barplot( x = 'day', y = 'total_bill', data = tips)
plt.show()
```

In the above graph, the y axis represents the mean 

```python
sns.barplot( x = 'day', y = 'total_bill', data = tips, estimator = np.mean)
plt.show()
```

```python
sns.barplot( x = 'day', y = 'total_bill', data = tips, estimator = np.mean, hue = 'sex')
plt.show()
```

```python
# what is preferred, lunch or dinner
sns.barplot( x = 'time', y = 'total_bill', data = tips)
plt.show()
```

```python
sns.barplot( x = 'day', y = 'total_bill', data = tips, hue = 'time')
plt.show()
```

```python
# Do those whose smoke, give more tip or more spent?
sns.barplot(x = 'smoker', y ='tip', data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

lmplot actually plot one more line (linear regression line) extra than scatterplot.

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips)
plt.show()
```
