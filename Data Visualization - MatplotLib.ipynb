# Bar Graph :-

```python
x= np.arange(5)
car1= [10,30,45,2,67]
car2= [34,12,30,12,8]
plt.bar(x,car1)
plt.show()
plt.bar(x,car2)
#By default the width of bar is 0.8 but we can chage also.
```

```python
plt.bar(x,car1, width = 0.5, tick_label = [2010,2011,2012,2013,2014])
plt.bar(x,car2, width = 0.5)
plt.show()
```

```python
plt.bar(x,car1, width = 0.5, tick_label = [2010,2011,2012,2013,2014],label = "car 1")
plt.bar(x+0.4,car2, width = 0.5, label = "Car 2")
#plt.xlabel("Year")
plt.ylabel("Sales in Million USD")
plt.title("Sales of Cars")
plt.legend()
#Changing x and y upper and lower limits:
plt.xlim(-1,6)
plt.ylim(-2, 90)
plt.show()
```

### Histograms :

```python
#Standard Normal Distribution (mean =0, std deviation =1)
x_snd = np.random.randn(500)
#To change mean and std dev of x_snd dataset :-
Hindi_marks = x_snd*7 + 80 
#This will make the data to have mean = 80 and std dev = 7
Hindi_marks
```

```python
Hindi_marks= np.round(Hindi_marks)
plt.hist(Hindi_marks)
plt.show()
```

Bins are the number of bars or width of each interval  you see in histogram. (By default it is 10)

```python
plt.hist(Hindi_marks, bins = 20)
plt.show()
```

```python
Hindi_marks.min()
```

```python
Hindi_marks.max()
```

```python
Hindi_marks.mean()
```

In each interval, lower limit value is included in bar  but upper limit value is not included in that bar.

# Creating Subplots(Multiple graphs in one):

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2= x**3
y3 = np.sin(x)
y4 = np.cos(x)
```

```python
#plt.subplot(no_of_rows,no_of_colums,serial_numberof_index)
plt.subplot(2,3,1)
plt.plot(x,y1)
plt.subplot(2,3,2)
plt.plot(x,y2)
plt.subplot(2,3,3)
plt.plot(x,y3)
plt.subplot(2,3,5)
plt.plot(x,y4)
plt.show()
```

# Plot 3D Graphs:-
np.meshgrid() returns coordinate matrices from coordinate vectors

```python
a = np.arange(3)
b = np.arange(4,11)
```

```python
np.meshgrid(a,b)
#It expands 1st array row-wise 
# by repeating to the (number_of_elements_in_2nd_array) times and 2nd array 
#colums wise i.e, vertically  by repeating the (number_of_elements_in_1st_array) times.
```

```python
a,b = np.meshgrid(a,b)
#Assigning
```

```python
fig = plt.figure()
#gca = get current access
fig.gca(projection = '3d')
plt.show()
```

# [Q]  Why 3D Plot is not showing for below cells?

```python
ax = fig.gca(projection = '3d')
#plot_surface takes parameters like what should be on x,y and z axes.
ax.plot_surface(a,b, a+b)
plt.show()
```

```python
ax.plot_surface(a,b,np.array([[1]]))
#This changes the elevation or heigt on z-axis.
plt.show()
```
